```python
import pandas as pd
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset=dataset.drop('User ID',axis=1) #axis=0: Refers to the rows /  axis=1: Refers to the columns
```

```python
independent=dataset[['Age','EstimatedSalary','Gender_Male']]
```

```python
dependent=dataset[['Purchased']]
```

```python
dataset['Purchased'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(input_train,output_train)
```

```python
predict=classifier.predict(input_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(output_test,predict)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf=classification_report(output_test,predict)
```

```python
print(clf)
```
